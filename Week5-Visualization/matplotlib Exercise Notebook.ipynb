```python

# DO NOT MODIFY

# import appropriate libraries

import matplotlib.pyplot as plt
import numpy as np
import sqlite3
import pandas as pd 
%matplotlib inline
```

```python

# DO NOT MODIFY

# we will use this dataset for some portions of this exercise.
# source: https://www.kaggle.com/hugomathien/soccer

def get_data():
    cnx = sqlite3.connect('/Users/Jacqueline/Documents/PythonForDataScience/Week5-Visualization/database.sqlite')
    df = pd.read_sql_query("SELECT * FROM Player_Attributes", cnx)
    return df

df = get_data()
```

```python
#DO NOT MODIFY

# Let's see what is in our dataset
df.describe()
```

```python

# modify this cell

def line_plot(df, x, y):
    ### BEGIN SOLUTION
    updated = df.sort_values(by=[x])
    updated = updated[:50]
    plt.title(x + " VS " + y)
    plt.xlabel(x)
    plt.ylabel(y)
    xs = updated[x]
    ys =updated[y]
   

    plt.plot(xs,ys)
    plt.show
    
    
    ### END SOLUTION
```

```python
# DO NOT MODIFY

# your function should give a plot similar to the following:
line_plot(df, 'potential', 'overall_rating')
```

```python
# modify this cell

def plot_histogram(df, X):
    ### BEGIN SOLUTION
    xs = df[X]
    xs.dropna(axis="index")
    plt.title("Histogram of " + X)
    plt.xlabel(X)
    plt.ylabel('# of players')
    plt.hist(xs)
    
    ### END SOLUTION
```

```python
# DO NOT MODIFY

# your plot should look similar to the following:
plot_histogram(df, 'overall_rating')
```

```python

# modify this cell

def plot_scatter(df, x, y):
    ### BEGIN SOLUTION
    xs = df[x]
    ys = df[y]
    plt.scatter(xs,ys)
    
    
    
    
    ### END SOLUTION
```
