```python
count = 5000

search_results = twitter_api.search.tweets(q=q, count=count)

statuses = search_results['statuses']
```

```python
# DO NOT MODIFY

all_text = []
filtered_statuses = []
for s in statuses:
    if not s["text"] in all_text:
        filtered_statuses.append(s)
        all_text.append(s["text"])
print(statuses[:1])
```

```python
retweets = None
### BEGIN SOLUTION
retweets = []
for tweet in filtered_statuses:
    retweet_count = tweet['retweet_count']
    text = tweet["text"]
    tweettup = text, retweet_count
    retweets.append(tweettup)

### END SOLUTION
```

```python
from operator import itemgetter
popular_tweets = None
### BEGIN SOLUTION
popular_tweets = sorted(retweets, key=itemgetter(1), reverse=True)[:10]
popular_tweets
### END SOLUTION
```

```python

assert len(popular_tweets) == 10, "Find the 10 most popular"
assert popular_tweets[0][0] >= popular_tweets[-1][0], "Make sure you are sorting in descending order"
```
